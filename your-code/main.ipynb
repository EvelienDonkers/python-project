```python
#Import Libraries
import random
import time

#Define classes
#Classes of colours for DARK background
class color:
    PURP = '\033[36m'
    BLUE = '\033[94m'
    GREY = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDER = '\033[4m'
    END = '\033[0m'

    

#Define all rooms and items variables

couch = {
    "name": "couch",
    "type": "furniture",
}

door_a = {
    "name": "door a",
    "type": "door",
}

key_a = {
    "name": "KEY for Door A",
    "type": "key",
    "target": door_a,
}

piano = {
    "name": "piano",
    "type": "furniture",
}

game_room = {
    "name": "game room",
    "type": "room",
}

outside = {
  "name": "outside"
}

bedroom1 = {
    "name": "bedroom 1",
    "type": "room",
}

bedroom2 = {
    "name": "bedroom 2",
    "type": "room",
}

living = {
    "name": "living room",
    "type": "room",
}

queenbed = {
    "name": "queen bed",
    "type": "furniture",
}

doublebed = {
    "name": "double bed",
    "type": "furniture",
}

dresser = {
    "name": "dresser",
    "type": "furniture",
}

table = {
    "name": "table",
    "type": "furniture",
}

door_b = {
    "name": "door b",
    "type": "door",
}

door_c = {
    "name": "door c",
    "type": "door",
}

door_d = {
    "name": "door d",
    "type": "door",
}

key_b = {
    "name": "KEY for Door B",
    "type": "key",
    "target": door_b,
}

key_c = {
    "name": "KEY for Door C",
    "type": "key",
    "target": door_c,
}

key_d = {
    "name": "KEY for Door D",
    "type": "key",
    "target": door_d,
}

master_key = {
    "name": "master key",
    "type": "key",
    "target": "all",
}

dice = {
    "name": "dice",
    "type": "furniture",
}

all_rooms = [game_room, bedroom1, bedroom2, living, outside]

all_doors = [door_a, door_b, door_c, door_d]

# define which items/rooms are related

object_relations = {
    "game room": [couch, piano, door_a, dice],
    "piano": [key_a],
    "outside": [door_d],
    "door a": [game_room, bedroom1],
    "bedroom 1": [queenbed,door_a,door_b,door_c],
    "door b": [bedroom1, bedroom2],
    "door c": [bedroom1, living],
    "door d": [living, outside],
    "queen bed": [key_b],
    "dresser": [key_d],
    "double bed": [key_c],
    "bedroom 2": [dresser,doublebed,door_b],
    "living room": [table,door_d,door_c],
    "dice" : [master_key]
}

# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_room": game_room,
    "keys_collected": [],
    "target_room": outside
}
```

```python
#Define functions:

#Prints a linebreak to space user input window from previous printing
def linebreak():
    """
    Print a line break
    """
    print("\n\n")

#Prints a countdown timer for the beginnig of the game:
def countdown(t=3):
    while t != 0:
        mins, secs = divmod(t, 60)
        timeformat = "{:02d}:{:02d}".format(mins, secs)
        print(color.RED + color.BOLD + timeformat + color.END, end="\r")
        time.sleep(1)
        t -= 1
    timeformat = "{:02d}:{:02d}".format(0,0)
    print(color.RED + color.BOLD + timeformat + color.END)
    

#Defines the outcomes of rolling the dice, which can lead to obtain master key
def dice_game(faces = 18):
    """
    Roll the dice
    """
    usr_choice = input(color.RED + "\nIt does not look like a normal dice..." + color.BOLD + "Type 'YES' if you want to roll it, press ENTER otherwise: ").strip().lower()
    if usr_choice == "yes":
        #print(color.RED + color-BOLD + "The mystery dice is rolling" + color.END)
        time.sleep(3)
        rolled_num = random.randint(1,faces)
        if rolled_num % 2 == 0:
            print(color.YELLOW + "\nThe outcome is ", rolled_num, " ...WOW a shiny golden key appeared out of nowere!" + color.END)
            return True 
        else:
            print(color.BOLD + "\nThe outcome is ", rolled_num, " ...Maybe you should roll it again..."+color.END)
            return False
    else:
        print(color.RED + "\nAren't you curious?" + color.END)
        return False
    
#Describes setting and initiate g# your code
def start_game():
    """
    Start the game
    """
    print(color.RED+color.BOLD+"You wake up on a COUCH and find yourself in a strange house with no windows which you have never been to before. You don't remember why you are here, but you suddenly hear a terrifying scream coming from the floor below your feet... You must get out of here, NOW!"+color.END)
    time=countdown()
    play_room(game_state["current_room"])
    

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        time.sleep(3)
        print(color.YELLOW + color.BOLD + "\nCONGRATS! YOU ESCAPED FROM AN ATROCIOUS DEATH!" + color.END)
    else:
        print(color.PURP + color.BOLD + "\nCurrent Room: " + room["name"].upper() + color.END)
        intended_action = input(color.BOLD + "\nType '1' to EXPLORE the room OR type '2' to EXAMINE an object! "+color.END).lower().strip()  
        if intended_action == "1":
            explore_room(room)
            play_room(room)
        elif intended_action == "2":
            examine_item(input(color.BOLD + "\nWhat would you like to EXAMINE? " + color.END).strip().lower())
        else:
            print(color.BOLD + "\nThat doesn't mean anything! TYPE '1' to EXPLORE the room or '2' to EXAMINE and object... if you want to save your life!" + color.END)
            play_room(room)
        linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    time.sleep(1)
    print(color.BOLD+"\nYou EXPLORE the room and you FIND: "+color.BLUE+color.UNDER+", ".join(items).upper() + color.END)
    #This is " + room["name"].upper()
    
def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = color.BOLD + "You examined " + item_name.upper() + ". " + color.END
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    #Master key has target all doors
                    if(key["target"] == item or key["target"]=="all"):
                        have_key = True
                if(have_key):
                    output += color.YELLOW + "You unlock it with a key you have." + color.END
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += color.RED + "It is locked but you don't have the key." + color.END
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    #Here inserted dice game outcomes
                    if item["name"] == "dice":
                        if dice_game() == True:
                            item_found = object_relations[item["name"]].pop()
                            game_state["keys_collected"].append(item_found)
                            output += color.YELLOW + color.BOLD + "You FOUND " + item_found["name"] + "." + color.END
                        else:
                            output += color.RED + "Nothing happend, but there is something strange..." + color.END
                    else:
                        item_found = object_relations[item["name"]].pop()
                        game_state["keys_collected"].append(item_found)
                        output += color.YELLOW + color.BOLD + "You FOUND " + item_found["name"] + "." + color.END
                else:
                    output += "There isn't anything interesting about it. You hear an indistinct tapping right behind you... "
            time.sleep(1)
            print("\n" + output)
            break

    if(output is None):
        print("\nSorry,the item you requested is not in the ro.." + color.RED + " Wait... What was that metallic noise? HURRY UP!"+color.END)
    
    if(next_room and input(color.BOLD + "\nType 'YES' if you want to move to the next room, press ENTER otherwise: " + color.END).strip().lower() == 'yes'):
        time.sleep(2)
        play_room(next_room)
    else:
        print(color.BOLD + "\nYou are still in the same room..." + color.END)
        play_room(current_room)
        
```
